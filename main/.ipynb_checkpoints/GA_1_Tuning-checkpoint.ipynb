# DO Tunning

```python
st_weights = {
    "p_margin": 0.30, # use fuel burnup coef instead of this
    "k_fa_max": 0.10,
    "k_left_right": 0.10, # modify in order to estimate this symmetry more accurate 
    "k_sym": 0.50
}
```

```python
#* updating weights
def update_weight(
    st_weights: dict, 
    variable_key: str,
    learning_rate: float = 0.05,
    steps: int = 5,
    importance: dict = {
        "k_fa_max": 0.0,
        "k_left_right": 0.20,
        "k_sym": 0.80
    }
    
):
    variable = st_weights[variable_key]
    incremental_weights = st_weights.copy()
    del incremental_weights[variable_key]

    incr_val = learning_rate
    grid = []
    new_weights = {}
    for i in range(steps+1):
        if i > 0:
            for incr_weight in incremental_weights.keys():
                incremental_weights[incr_weight] += incr_val*importance[incr_weight]
        grid.append(
            {
                variable_key: variable,
                **incremental_weights.copy()}
            
        )
        variable -= learning_rate
    return grid
```

```python
weights_grid = update_weight(
    st_weights,
    "p_margin"
)
weights_grid
```

```python
from concurrent.futures import ThreadPoolExecutor, ProcessPoolExecutor, as_completed
```

```python
with ProcessPoolExecutor(max_workers=len(weights_grid)) as executor:
    future_weight_grid = []

    for n, i in enumerate(weights_grid):
        ga = GA.no_fuel_mask(
            ancestor_0.loc[p_cols].to_numpy(), 
            ancestor_0.loc[t_cols].to_numpy(),
            #* refuel_only provided
            # refuel_only=[0, 10],
            population_size=200,
            mate_probability=0.6,
            permutation_mutation_probability=0.3,
            mutation_probabilty_fresh_fuel=0.4,
            fintess_weights = i
        )
        st_time = time.time()
        future_weight_grid.append(executor.submit(ga.search, generations=100))
    for n, future in enumerate(as_completed(future_weight_grid)):      
        trained_population = future.result()
        fn_time = time.time()
        best_ten = trained_population[:5]
        weights_grid[n]["ga_results"] = best_ten.copy()
        weights_grid[n]["exec_time"] = fn_time - st_time
        print(weights_grid[n])
    
```

```python
# estimation of permutation number and getting step by step refueling actions
for res in weights_grid:
    for core in res["ga_results"]:
        # print(core["id"])
        cycles = refueling_pattern.find_pattern(
            ancestor_chromosome["fuels_gnome"],
            core["fuels_gnome"]   
        )
        pattern = refueling_pattern.decode_pattern(ancestor_chromosome["fuels_gnome"], cycles)
        core["permutations"] = pattern
```
