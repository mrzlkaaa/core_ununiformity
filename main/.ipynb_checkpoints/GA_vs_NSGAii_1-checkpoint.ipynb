```python
# separate by gpi
sga_gpis = defaultdict(list)
nsgaii_gpis = defaultdict(list)

for i in folder_list:
    path = os.path.join(
        folder,
        i
    )
    for gpi in legend:
        if "nsgaii" in i:
            if str(gpi) in i:
                nsgaii_gpis[gpi].append(
                    path
                )
        else:
             if str(gpi) in i:
                sga_gpis[gpi].append(
                    path
                )
        
```

```python
def sepbyba(d, dct):
    gbis_ba = {}
    for gpi in legend:
        gbis_ba[gpi] = defaultdict(list)
        for l in d[gpi]:
            if "best" in l:
                data = load(l)
                if dct:
                    data = [i["fitness_score"] for i in data]
                gbis_ba[gpi]["best"].append(
                    data
                )
            else:
                gbis_ba[gpi]["aver"].append(
                    load(l)
                )
    return gbis_ba
```

```python
sga_gpis_ba = sepbyba(sga_gpis, True)
nsgaii_gpis_ba = sepbyba(nsgaii_gpis, False)
```

```python
def scaling_factor(Fval):
    # print(1 if Fval >= 0.945 else np.exp(20 * (Fval - 0.95) ))
    return 1 if Fval >= 0.943 else np.exp(50 * (Fval - 0.943) )
```

```python
def get_difef(arr):
    rates = []
    for i in arr:
        rate = np.array([])
        for j in range(len(i)-1):
            rate = np.append(
                rate,
                abs( ( i[j+1] - i[j] )) #  / i[j+1] ) * 100
            )
        rates.append(rate)
    return rates
```

```python
def eff(difef, tr):
    return difef / tr
```

```python
x_lim = 1.2
x = np.linspace(1.2, 2, 50)
```

```python
y = x.copy()
# pos, val
stage = [None, None]
for i in range(len(x)):
    if x[i] <= 1.5:
        y[i] = 1 / ( x[i]/x_lim )
        continue
    elif stage[0] is None:
        print("assigning")
        stage[0] = x[i-1]
        stage[1] = y[i-1]
    y[i] = stage[1] * np.exp(-5* x[i] ) / np.exp ( -5* stage[0] )
```

```python
sns.scatterplot(x=x, y=y)
```
